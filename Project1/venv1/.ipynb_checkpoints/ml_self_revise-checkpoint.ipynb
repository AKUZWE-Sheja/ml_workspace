```python
Q1 = df['selling_price'].quantile(0.25)
Q3 = df['selling_price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['selling_price'] < lower_bound) | (df['selling_price'] > upper_bound)]
print(outliers)
```

```python
print("The 1st line calcuates Q1 which is 25%")
print("The 2nd line calcuates Q3 which is 75%")
print("A larger IQR indicates more variability in the middle 50% of the data.")
print("The factor of 1.5 is a common choice that helps to identify extreme values.")
print("Any point < lower bound or > upper bound is outling.")
```

```python
# Data
data = [140, 189, 185, 195, 149, 189, 147, 154, 174, 169, 195,
        159, 192, 155, 191, 153, 157, 140, 144, 140, 172, 157,
        181, 182, 166, 167, 140, 140]

# Finding unique values and counts
unique_values, counts = np.unique(data, return_counts=True)

# Creating DataFrame
df = pd.DataFrame(data)

# Count of data points
n = df.count()

# Finding the number of bins
k = int(np.ceil(np.log2(n) + 1))

# Plotting the histogram
plt.figure(figsize=(23, 3))
plt.hist(data, bins=k, edgecolor="yellow", color='blue', alpha=0.3)

# Setting x-ticks to unique values if needed
plt.xticks(unique_values)

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data')

# Display the plot
plt.show()
```
